```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.arange(5)
y=(20,35,30,35,27)
plt.bar(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('scatter plot')
plt.show()
```

```python
plt.plot(x,y)
plt.show()
```

```python
x=np.linspace(0,20,100)
print(x)
y=np.sin(x)
print(y)

plt.plot(x,y,label='sample label')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('sample plot title')
plt.savefig('new_plot.jpg')
plt.show()
```

```python
x=np.linspace(0,20,100)
y=np.sin(x)
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(1,1,1)
ax.plot(x,y)
```

```python
ax.plot(x,y,'b-',linewidth=2,label='sample label')
```

```python
x1=np.linspace(0,100,20)
x2=np.linspace(0,100,20)
x3=np.linspace(0,100,20)

y1=np.sin(x1)
y2=np.cos(x2)
y3=np.tan(x3)

fig,ax = plt.subplots(figsize=(8,4))

ax.plot(x1,y1,label='sin')
ax.plot(x2,y2,label='cos')
ax.plot(x3,y3,label='tan')

ax.legend(loc='best',prop={'size':'large'})
```

```python
x1=np.linspace(0,100,20)
x2=np.linspace(0,100,20)
x3=np.linspace(0,100,20)

y1=np.sin(x1)
y2=np.cos(x2)
y3=np.tan(x3)

fig,(ax1,ax2,ax3) = plt.subplots(nrows=3,ncols=1,sharex=False,sharey=False,figsize=(12,6))

ax1.plot(x1,y1,label='sin')
ax2.plot(x2,y2,label='cos')
ax3.plot(x3,y3,label='tan')

ax1.grid(True)
ax1.grid(True)
ax1.grid(True)

fig.suptitle('A simple Multi axis line plot')
```
